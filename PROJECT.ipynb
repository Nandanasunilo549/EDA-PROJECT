# Used cars dataset

 ### The dataset represents information about 1000 used bikes listed for sale. Each record includes details such as the bike's Brand, Model, Manufacturing Year, Price (USD), and Condition (e.g., New, Good, Fair). Additional columns provide insights into Mileage, Fuel Type, Engine Capacity (cc), and Color, along with the Location and Seller Type (Individual/Dealer). It also includes Date Listed, a brief Comments field describing the bike, and its Status (Available/Sold). This dataset is ideal for analyzing trends in the second-hand bike market, such as pricing, popular brands, and demand by condition.
